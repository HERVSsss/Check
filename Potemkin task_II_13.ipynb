#### Допишите следующий код на Python так, чтобы он генерировал матрицу A (состоящую из 0 и 1) раз-
#### мера 15×28, показанную на Рис. 1 (для визуализации матрицы использована функция plt .imshow(A)):

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from numpy import linalg
```

```python
a = np.zeros ((15 , 28))
a [2: -2 ,1] = 1; a [2 ,2:6] = 1
a [2:7 ,6] = 1; a [7: -2 ,7] = 1
a [7 ,2:7] = 1; a [ -3 ,2:7] = 1
a [2: -2 , 10] = 1; a [2: -2 , 14] = 1
a [2: -2 , 18] = 1; a [ -3,10:19] = 1
a [2: -2, 26] = 1; a[12, 21:-1] = 1
a [7, 21:-1] = 1; a[2, 21:-1] = 1
```

```python
plt.imshow(a)
```

#### Постройте SVD разложение матрицы A. Чему равен rank(A)?

```python
U, E, V = linalg.svd(a)
#with matrix_rank function from numpy
print('rank(a) = ', linalg.matrix_rank(a))
```

```python
Z = np.zeros((15,28))
Z[:16,:15] = np.diag(E)
E = Z
U = np.reshape(U, (15,15))
V = np.reshape(V, (28,28))

plt.imshow(U @ E @ V)
```

```python
for i in range(linalg.matrix_rank(a) + 1):
    B_i = U[:,:i] @ E[:i,:i] @ V[:i,:]
    fig = plt.figure()
    plt.imshow(B_i)
```
